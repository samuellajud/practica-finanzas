```python

def obtener_datos_ticker(ticker):
    tker = ticker
    tesla = yf.Ticker(tker)
    tesla_info = tesla.info
    historico_tesla = tesla.history(period="max")
    
    # Mostrar información del ticker
    print("Información del Ticker:")
    print(f"TICKER: {tesla_info['symbol']}")
    print(f"Sector: {tesla_info['sector']}")
    print(f"Web: {tesla_info['website']}")
    print()
    
    # Mostrar DataFrame del histórico
    print("Histórico del Ticker:")
    print(historico_tesla.tail())
    print()
    
    # Mostrar los primeros 10 valores de la columna 'Close'
    print("Primeros 10 valores de 'Close':")
    print(historico_tesla["Close"].tail(10))
    print()
    
    # Resetear el índice para graficar correctamente
    historico_tesla.reset_index(inplace=True)
    
    # Graficar los datos de cierre
    historico_tesla.plot(title=tker, x="Date", y="Close")
    plt.show()

# Solicitar al usuario el ticker
ticker_input = input("Ingrese el nombre del ticker: ")
obtener_datos_ticker(ticker_input)
```
